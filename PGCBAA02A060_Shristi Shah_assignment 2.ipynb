```python
#checking what happens if range is beyond numbe rof columns
#positional indexers are out-of-bounds
extra_range =list(range(9))
df.iloc[:,extra_range]
```

```python
#subsetting rows for all columns using range
extra_range =list(range(9))
df.iloc[extra_range,:]
```

```python
#range with 3 paramters
select_range=list(range(0,6,3))
df.iloc[select_range,:]
```

```python
#range for both rows and columns
df.iloc[:,2:]#all rows,all columns starting from column 3
df.iloc[:,:5]#all rows,all columns upto column 4
```

```python
#slice rows 3 to 8 inclusive
df.iloc[3:9,:]
```

```python
#slice every 10th row starting from 0 to 100
df.iloc[0:100:10,:]
```

```python
#whatif one of the 3 index in range is missing
df.iloc[:100:10,:6:2]
df.iloc[0::10,0::2]
df.iloc[0:100:,0::]
df.iloc[0:100:,::]
df.iloc[0:100:,::3]
```

```python
#finding grouped mean based on continent
df.sortby()
```

```python
#finding datatype
type(df.groupby('continent'))
type(df.groupby('continent')['lifeExp'])
```

```python
#grouping data based on multiple columns and having multiple means
df.groupby(['year','continent'])[['lifeExp','pop']].mean()
```

```python
#to display year value against all continents using reset_index
df.groupby(['year','continent'])[['lifeExp','pop']].mean().reset_index()
```

```python
#get count of unique values againt each filter using nunique 
df.groupby(['year','continent'])['gdpPercap'].nunique()
```

```python
#get count of unique values againt each filter using value_counts
#get the value with the count of occurrence
df.groupby(['year','continent'])['gdpPercap'].value_counts()
```

```python
#data visualization using plot method
df.groupby('year')['gdpPercap'].mean().plot()
```
